### Normalization
1.Normalization is conducted to make features value ranges from 0 to 1.
          x=[x-min(x)/{max(x)-min(x)}]<br>
2.This whole process is also know as MinMaxScaler

```python
#1st method:-
#By calculating through formula
def normalize(f5):
    result=f5.copy()
    for i in df.columns:
        max_value=df[i].max()
        min_value=df[i].min()
        result[i]=(df[i]-min_value)/(max_value-min_value)
    return result
    
```

```python
df1=normalize(df)
df1
```

```python
#2nd method:
#By using the library of python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
df2=scaler.fit_transform(df)
```
